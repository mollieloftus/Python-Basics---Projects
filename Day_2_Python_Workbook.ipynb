```python
count = 0
max_count = 5
user_input = ''

print("Starting loop: Enter 'stop' to quit or let count reach 5.")

while count < max_count and user_input != 'stop':
  print(f"Current count: {count}")
  user_input = input("Enter 'stop' to exit or anything else to continue: ")
  count += 1

print("Loop finished.")

```

```python
password = 123
user_entry = int(input("Enter Password: "))

attemps = 1 #setting counter

while password != user_entry:
  print("Password is incorrect. Entry denied.")
  user_entry = int(input("Enter Password: "))

  attemps += 1
  if attemps > 2:
    break
if password == user_entry:
  print("Password is correct. Entry granted.")
else:
  print("Your account has been locked.")
```

```python
secret_number = 8
tries_left = 3

while tries_left > 0:
    guess = int(input("Guess the secret number (1-10): "))

    if guess == secret_number:
        print("ðŸŽ‰ Correct! You win!")
        break

    elif guess < secret_number:
        print("Too low.")

    else:
        print("Too high.")

    tries_left -= 1 #add attemp
    print(f"Tries left: {tries_left}")

if tries_left == 0:
    print("ðŸ˜¢ Game over. The correct number was", secret_number)
```
